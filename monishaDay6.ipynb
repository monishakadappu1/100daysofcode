```python
import pandas as pd
# The data set file is in the same Github folder from where you got this file 
percentage_people_below_poverty_level = pd.read_csv('PercentagePeopleBelowPovertyLevel.csv', encoding="windows-1252")

percentage_people_below_poverty_level.head()

percentage_people_below_poverty_level.replace('-',0.0,inplace = True)
#print(type(percentage_people_below_poverty_level))
percentage_people_below_poverty_level.head(5)
percentage_people_below_poverty_level=percentage_people_below_poverty_level.astype({'poverty_rate':float})
percentage_people_below_poverty_level.dtypes

area_group = percentage_people_below_poverty_level.groupby('Geographic Area')
print((area_group.get_group('AL').poverty_rate.sum())/585)
```

```python
# Code here to clean the data because some fields have "-" character. Clean it with floating value of zero

# Code here to convert the data into floating type

area_list = list(percentage_people_below_poverty_level['Geographic Area'].unique())
area_poverty_ratio = []

# run a for loop and populate the area_poverty_ratio by calculating the average poverty across each state
# Eg If you take AL as state, there are 585 cities in AL and sum of all poverty rate / 585 will be your poverty rate of AL
for i in area_list:
    x = area_group.get_group(i).poverty_rate.sum()
    area_poverty_rate = x/area_group.get_group(i).poverty_rate.count()
    area_poverty_ratio.append(area_poverty_rate)
dataarea = pd.DataFrame({'area_list': area_list,'area_poverty_ratio':area_poverty_ratio})


dataarea = dataarea.sort_index(ascending = False)
#print(dataarea)

# sort the data using the descending ordered index values
# code here


# visualization
plt.figure(figsize=(15,10))
#code here and show how creative you are
plt.plot(dataarea['area_list'],dataarea['area_poverty_ratio'],label = 'ratio')
sns.set()
plt.xlabel('area')
plt.ylabel('area_poverty_ratio')
plt.title('Poverty Rate Given States')
plt.legend()
plt.show()
#sns.lmplot(x='area_list',y='area_poverty_ratio',data=dataarea)
```
