```python
d_x = np.linspace(-3,3,8)
```

```python
d_y = np.linspace(-5,5,9)
```

```python
d_xl,d_yl = np.meshgrid(d_x,d_y)
```

```python

z = np.sin(d_xl) + np.sin(d_yl)
```

```python
print(z)
```

```python
plt.plot(z)
```

```python
plt.title('Plot for sin(dx) + sin(dy)')
```
