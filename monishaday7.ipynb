```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set(color_codes=True)
```

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.lmplot(y ='tip',x='total_bill',data = tips,scatter=False)
```

```python
g = sns.lmplot(x = 'total_bill',y = 'tip',hue = 'smoker',data=tips,scatter=False)
```

```python
g = sns.lmplot(x = 'total_bill',y = 'tip',hue = 'smoker',col = 'time',data=tips,scatter=False)
```

```python
g  = sns.lmplot(x = 'total_bill',y = 'tip',hue = 'time',col = 'smoker',data=tips,scatter=False)
```

```python
g = sns.scatterplot(x = 'total_bill',y = 'tip',hue = 'size', data=tips)
```

```python
#current =sns.cubehelix_palette(8,rot = -0.75,start = .5)
current = (sns.color_palette("cubehelix", 8))
g =sns.scatterplot(x = 'total_bill',y = 'tip',hue = 'size',palette =sns.cubehelix_palette(rot = -0.75,start = .5),  data=tips)
```

```python

g =sns.scatterplot(x = 'total_bill',y = 'tip',hue = 'size',palette =sns.cubehelix_palette(rot = -0.75,start = .5),  data=tips,size = 'size')
```

```python
g = sns.relplot(x = 'total_bill',y = 'tip',hue = 'time',data = tips,kind = 'line')
```
