```python
countries = ['Argentina', 'USSR', 'Germany', 'China', 'Japan', 'USA']
obj = dict(zip(countries,ww2_dict.values()))
```

```python
print(obj)
```

```python
ww2_series = Series(obj)
```

```python
print(ww2_series)
```

```python
print(ww2_series + obj)
```
